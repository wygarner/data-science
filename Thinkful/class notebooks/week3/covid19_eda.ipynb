```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline


covid19 = pd.read_csv("data/2019-nCoV-cases-JHU.csv")

covid19["Date"] = pd.to_datetime(covid19["Date"])
covid19.loc[covid19["Region"] == "United States", "Region"] = "US"

covid19.info()
```

```python
covid19["weekday"] = covid19["Date"].dt.weekday
covid19["weekday_name"] = covid19["Date"].dt.weekday_name

# fmt: off
(covid19[["weekday", "weekday_name"]]
  .drop_duplicates()
  .sort_values("weekday"))
# fmt: on
```

```python
px.scatter(covid19, x="Date", y="Confirmed", color="Region")
```

```python
state_map = {
    "Washington": "WA",
    "Illinois": "IL",
    "California": "CA",
    "Arizona": "AZ",
    "Chicago": "IL",
}

us = covid19[covid19["Region"] == "US"].copy()

us["state"] = us["Province"].str.split(", ").str[-1]
us["state"] = us["state"].str.split(" \(").str[0]

us = us[~(us["state"] == "Unassigned Location")]
us["state"] = us["state"].replace(state_map)
us.head()
```

```python
# fmt: off
us_agg = (us[["Date", "state", "Confirmed"]]
 .groupby(["Date", "state"])
 .sum()
 .reset_index())
# fmt: on
```

```python
px.scatter(us_agg, x="Date", y="Confirmed", color="state")
```
