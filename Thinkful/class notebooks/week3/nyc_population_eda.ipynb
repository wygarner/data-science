```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

# connect to database and read in last quarters financials
# financials_df = pd.read_sql('awesome query', engine)

data_url = "https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/nyc-population-tall.csv"
nyc_pop = pd.read_csv(data_url)
nyc_pop.head()
```

```python
# year | manhattan | bronx  | ....
# 1900 | 1850093   | 200507 | ....
```

```python
nyc_wide = pd.pivot(data=nyc_pop, index="year", columns="borough", values="population")
nyc_wide
```

```python
nyc_pop.columns = ["Year", "Borough", "Population"]

plt.figure(figsize=(10, 5))
sns.lineplot(x="Year", y="Population", hue="Borough", data=nyc_pop)
plt.title("NYC Population 1900 - 2010 by Borough")

plt.savefig("nyc_pop_lineplot.png")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
for c in nyc_wide.columns:
    plt.plot(nyc_wide.index, nyc_wide[c], label=c)

plt.title("NYC Population 1900 - 2010 by Borough")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(loc="lower right")

plt.savefig("nyc_pop_lineplot.png")
plt.show()
```
