#### Are there any differences between the genders, ethnicities, and parental level of education with respect to their performances in exams?

```python
female = education_df[education_df["gender"] == "female"]
male = education_df[education_df["gender"] == "male"]
```

```python
for elem in ["math score", "reading score", "writing score"]:
    plt.hist(female[elem], alpha=0.5, label="female " + elem)
    plt.hist(male[elem], alpha=0.5, label="male " + elem)
    plt.axvline(female[elem].mean(), c="blue")
    plt.axvline(male[elem].mean(), c="orange")
    plt.legend()
    plt.show()
```
