```python
life_ex = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/life_expectancy.csv')
```

```python
life_ex.head()
```

```python
ids = ['Country Name','Country Code', 'Indicator Name','Indicator Code']
melt_fields = set(life_ex.columns) - set(ids)
melted = pd.melt(life_ex, id_vars=ids,
                value_vars=melt_fields,
                var_name='Year', value_name='Value')
```

```python
melted.info()
```

```python
melted
```

```python
melted['Indicator Name'].value_counts()
```

```python
melted['Indicator Code'].value_counts()
```

```python
melted=melted.drop(columns='Indicator Name')
melted=melted.drop(columns='Indicator Code')
melted=melted.drop(columns='Country Code')
```

```python
dropped = melted.dropna()
```

```python
melted['Value2']=np.nan
for country in melted['Country Name'].unique():
    melted['Value2'][melted['Country Name']==country]= melted['Value'][melted['Country Name']==country].interpolate()
```

```python
melted['Value3']=np.nan
for year in melted['Year'].unique():
    melted['Value3'][melted['Year']==year]= melted['Value'][melted['Year']==year].median()
```
