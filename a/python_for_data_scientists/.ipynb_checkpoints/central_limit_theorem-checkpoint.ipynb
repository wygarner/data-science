```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
results = []
for i in range(1,1000):
    trials = np.random.poisson(5, i)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
plt.show()
```

```python
means= []
for i in range(1000):
    means.append(np.mean(np.random.poisson(5, 1000)))
plt.hist(means)
plt.show()
```

```python
results = []
for i in range(1,1000):
    trials = np.random.normal(5,5,i)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
plt.show()
```

```python
means= []
for i in range(1000):
    means.append(np.mean(np.random.poisson(5, 1000)))
plt.hist(means)
plt.show()
```
